```python
# імпортуємо бібліотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("menu.csv", encoding = 'utf-8') 

```

```python
# Перевіряємо чи є нульові значення в датафреймі
df.isnull().values.any() 
```

```python
# виводимо датафрейм
df
```

```python
# функція створення кольорованої таблиці
def ctable(value):
    color = 'black'
    try:
        # при значення більше 25 розфарбовуємо значення у червоний
        if value > 25:
            if value in df['Total Fat'].tolist():
                color = 'red'
        # при значенні менше 15 - у зелений
        elif value < 15:
            if value in df['Total Fat'].tolist():
                color = 'green'
    except Exception as e:
        print(f"An error occurred: {e}")
        color = 'black'
    return f'color: {color}'
# обираємо 10 перших значень з таблиці
s = df.head(10)[['Category', 'Item', 'Calories', 'Total Fat']]
# сортуємо по полю жирів та розфарбовуємо таблицю
s = s.sort_values(by='Total Fat', ascending=True).reset_index(drop=True).style.applymap(ctable, subset=['Total Fat'])
s
```

```python
def low_fat_calories(dataframe): #обираємо продукти з найменшою кількістю жирів та калорій
    new_df = df.sort_values(by=['Calories']).loc[(df['Total Fat']<10) & (df['Calories']<350 ),['Category','Item','Calories','Total Fat']]
    return new_df
low_fat_calories(df)
```

```python
mean_A = df['Total Fat'].mean() # шукаємо мат.очікування жирів
variance_A = df['Total Fat'].var() # шукаємо дисперсію жирів

print(f'Mean Total Fat: {mean_A}')
print(f'Variance Total Fat: {variance_A}')
```

```python
def high_calories_sugars(dataframe): # шукаємо позиції, де калорії більше 500, або сахар вище 100
    new_df = dataframe[(dataframe['Calories']>= 500) | (dataframe['Sugars']>= 100)]
    return new_df
high_calories_sugars(df)
```

```python
plt.figure(figsize=(12, 10)) # стовпчаста діаграма
sns.barplot(x='Calories',y='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 10)) # ліній на діаграма із дисперсією  калорійн в залежності від категорії
sns.lineplot(x='Category',y='Calories',data=df)
plt.show()
```

```python
def extreme(dataframe): #обираємо топ 10 продуктів з найбільшним холестерином та сахаром, виводимо лише парні андекси
    new_df1 = dataframe.sort_values(by='Cholesterol', ascending = False)[['Category','Item','Total Fat','Cholesterol','Sugars']]
    new_df2 = dataframe.sort_values(by='Sugars', ascending = False)[['Category','Item','Total Fat','Cholesterol','Sugars']]
    result_df = pd.concat([new_df1.head(10),new_df2.head(10)])
    return result_df.loc[result_df.index %2 == 0]
extreme(df)
```

```python
sns.stripplot(x = 'Calories', y = 'Category', data = df) # точковка діаграма та стовбчаста
print(extreme(df).plot.hist())
```

```python
parameters = ['Calories', 'Total Fat', 'Protein', 'Carbohydrates', 'Sodium'] # теплова мапа

plt.figure(figsize=(10, 6))
heatmap_data = df[parameters][0:8]
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5)
plt.title('Heatmap for Selected Products')
plt.show()
```

```python
import plotly.graph_objects as go # лінійна діаграма з анотаціями

df_random_20_products = df.sample(n=20,random_state=41)
df_top_20_calories = df.nlargest(20, 'Calories')

fig = go.Figure()
frame = df_top_20_calories
for i, column in enumerate(['Calories', 'Total Fat', 'Sodium', 'Protein']):
    # Додаємо лінію для кожної ознаки
    fig.add_trace(go.Scatter(
        x=frame['Item'],
        y=frame[column],
        mode='lines',
        name=column
    ))

    # Додаємо анотацію для середнього значення
    avg_value = frame[column].mean()
    max_index = df_top_20_calories[column].idxmax()
    if i == 3:
        df_removed_max = df_top_20_calories.drop(index=max_index)
        max_index = df_removed_max[column].idxmax()
    fig.add_annotation(
        x=frame['Item'][max_index],
        y=frame[column][max_index],
        text=f'Mean {column}: {avg_value:.2f}',
        showarrow=True,
        arrowhead=2,
        arrowcolor='black',
        arrowwidth=2,
    )

# Налаштовуємо вигляд графіку
fig.update_layout(
    title='Line Plot with Annotations',
    xaxis_title='Item',
    yaxis_title='Value',
    legend_title='Features'
)

# Показуємо графік
fig.show()
```
